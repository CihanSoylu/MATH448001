**Find gradients and update parameters**

In this step you will implement a function which updates the weights and biases according to the gradient of the loss function calculated for a mini-batch of $N$ data points. 
 
In order to update $w_k$, we first need to find the derivative of $J(y, \hat{y})$ with respect to $w_k$:
$$
\begin{eqnarray}
\frac{\partial J}{\partial w_k} & = & -\frac{1}{N}\sum_{i=1}^N\frac{\partial L(y^{(i)}, \hat{y}^{(i)})}{\partial w_k} \\
& = & -\frac{1}{N}\sum_{i=1}^N\frac{\partial L(y^{(i)}, \hat{y}^{(i)})}{\partial \hat{y}^{(i)}}\frac{\partial  \hat{y}^{(i)}}{\partial z^{(i)}} \frac{\partial z^{(i)}}{\partial w_k} \\
& = & -\frac{1}{N}\sum_{i=1}^N (y^{(i)} - \hat{y}^{(i)}) x^{(i)}_k
\end{eqnarray}
$$
Similarly the derivatives with respect to $b$:
$$
\begin{eqnarray}
\frac{\partial J}{\partial b} & = & -\frac{1}{N}\sum_{i=1}^N\frac{\partial L(y^{(i)}, \hat{y}^{(i)})}{\partial b} \\
& = & -\frac{1}{N}\sum_{i=1}^N\frac{\partial L(y^{(i)}, \hat{y}^{(i)})}{\partial \hat{y}^{(i)}}\frac{\partial  \hat{y}^{(i)}}{\partial z^{(i)}} \frac{\partial z^{(i)}}{\partial b} \\
& = & -\frac{1}{N}\sum_{i=1}^N (y^{(i)} - \hat{y}^{(i)})
\end{eqnarray}
$$

Now given a mini-batch $\{(x^{(i)}, y^{(i)})\}_{i=1}^N$, the update equations for the weights are
$$
\begin{eqnarray}
w_k^{new} & = & w_k^{old} - \alpha \frac{\partial J}{\partial w_k} = w_k^{old} + \alpha \frac{1}{N}\sum_{i=1}^N (y^{(i)} - \hat{y}^{(i)}) x^{(i)}_k \\
b^{new} & = & b^{old} - \alpha \frac{\partial J}{\partial b} = b^{old} + \alpha \frac{1}{N}\sum_{i=1}^N (y^{(i)} - \hat{y}^{(i)})
\end{eqnarray}
$$
