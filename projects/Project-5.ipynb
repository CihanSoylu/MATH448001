In this exercise you will create another classifer for Dogs vs. Cats task using 'transfer learning'. In the lectures you have seen that a convolutional network consists of a series of convolutional layers (together with pooling layers) followed by a series of densely connected layers. We think of the convolutional part (feature maps) as extracting the important features from the images and the dense part as classifying these features into categories. Transfer learning is a very powerful method which refers to using the feature maps (the convolutional layers) learned by a deep convolutional network on a different dataset as the feature maps for the problem at hand and training a classifier on top of these feature maps. In other words, instead of designing the convolutional layers from scratch, we use the convolutional layers of an already trained model. The advantage of this is that the pre-trained models are usually highly complex models trained on very big datasets using vast amount of computational power. Most features learned by these models can be transfered to the problems where the data is similar to the dataset that the pre-trained model is trained on. 

A very popular dataset these pre-trained models train on is called [ImageNet](http://www.image-net.org/). This dataset contains millions of images in thousands of categories. Using Tensorflow you can load a model with weights learned on ImageNet dataset and use the convolutional layers of this model as feature maps in your model and train a separate classifier.  

Let's load one of these pre-trained models. For a list of available models see this [link](https://www.tensorflow.org/api_docs/python/tf/keras/applications).

```
input_shape = (128,128,3)
pretrained_model = tf.keras.applications.DenseNet121(input_shape=input_shape,
                                                    include_top=False,
                                                    weights='imagenet')
```

Let's go over the arguments provided to this function. 

- `input_shape` is shape of the input data.
- `include_top` refers to the classifier on top of the convolutional layers. If False the classifier is not included. You will be training your own classifier for the dataset at hand so this will be false. 
- `weights` refers to the weights learned by training on the dataset provided in this argument. In the example above, the model will be loaded with the weights learned through training on ImageNet dataset.
